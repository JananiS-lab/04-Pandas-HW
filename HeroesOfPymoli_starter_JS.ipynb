```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
player_count= purchase_data['SN'].nunique()
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Unique_Items= purchase_data['Item ID'].nunique()
Average_price= round(purchase_data['Price'].mean(),2)
Purchases= purchase_data['Purchase ID'].count()
Revenue= purchase_data['Price'].sum()
summary_purchase_data= pd.DataFrame({"Number of Unique Items" : [Unique_Items],
                                   "Average Price" : [f'${Average_price}'],
                                   "Number of Purchases": [Purchases],
                                   "Total Revenue" :[f'${Revenue}']})
summary_purchase_data
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
pd.options.display.float_format = '${:,.2f}'.format
gender_count=purchase_data.groupby('Gender')['SN'].nunique()
percent_count=(gender_count/player_count)*100
final_result=pd.concat([gender_count.to_frame(), percent_count.to_frame()],axis=1,sort=False)
final_result.columns=['Total Count', 'Percentage of Players']
final_result
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pd.options.display.float_format = '${:,.2f}'.format
gender_price=purchase_data.groupby('Gender')['Price'].sum()
gender_average=round(purchase_data.groupby('Gender')['Price'].mean(),2)
gender_number=purchase_data.groupby('Gender')['SN'].count()
gender_avgspend=round((gender_price/gender_count),2)
gender_final=pd.concat([gender_number.to_frame(), gender_average.to_frame(), 
                        gender_price.to_frame(),gender_avgspend.to_frame()],axis=1,sort=False)
gender_final.columns=['Purchase Count', 'Average Purchase Price', 'Total Purchase Value', "Avg Total Purchase per Person"]
gender_final
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
pd.options.display.float_format = '{:,.2f}%'.format
bins =[0,9,14,19,24,29,34,39,100]
Ages=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
purchase_data["Age Groups"]=pd.cut(purchase_data['Age'],bins,labels=Ages, include_lowest = True)
gender_bins=purchase_data.groupby('Age Groups')['SN'].nunique()
percent_player= (gender_bins/player_count)*100
percent_player
Age_demographics=pd.concat([gender_bins.to_frame(), percent_player.to_frame()], axis=1,sort=False)
Age_demographics.columns=['Total Count', 'Percentage of Players']
Age_demographics
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pd.options.display.float_format = '${:,.2f}'.format
bins =[0,9,14,19,24,29,34,39,100]
Ages=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
purchase_data["Age Groups"]=pd.cut(purchase_data['Age'],bins,labels=Ages, include_lowest = True)
purchase_count=purchase_data.groupby('Age Groups')['Purchase ID'].count()
purchase_value=purchase_data.groupby('Age Groups')['Price'].sum()
Average_pur_price=(purchase_value/purchase_count)
Avg_ppp=(purchase_value/gender_bins)
purchasing_analysis=pd.concat([purchase_count.to_frame(), Average_pur_price.to_frame(), 
                               purchase_value.to_frame(), Avg_ppp.to_frame()], axis=1, sort=False)
purchasing_analysis.columns=['Purchase Count', 'Average Purchase Price', 
                             'Total Purchase Value', 'Avg Total Purchase per Person ']
purchasing_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pd.options.display.float_format = '${:,.2f}'.format
spenders=purchase_data.groupby('SN')['Price'].sum()
pur_count=purchase_data.groupby('SN')['Purchase ID'].count()
avg_pur_pr=(spenders/pur_count)
spenders_analysis=pd.concat([pur_count.to_frame(), avg_pur_pr.to_frame(), spenders.to_frame()],axis=1,sort=False)
spenders_analysis.columns=['Purchase Count', 'Average Purchase Price', 'Total Purchase value']
spenders_sorted=spenders_analysis.sort_values(['Purchase Count', 'Average Purchase Price', 'Total Purchase value'], 
                                       ascending=False).groupby('Total Purchase value')
spenders_sorted.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
pd.options.display.float_format = '${:,.2f}'.format
popular_table=purchase_data[["Item ID", "Item Name", "Price"]]
popular_count=purchase_data.groupby(['Item ID','Item Name'])["Purchase ID"].count()
popular_purval=purchase_data.groupby(['Item ID','Item Name'])["Price"].sum()
popular_price=(popular_purval/popular_count)
popularity_analysis=pd.concat([popular_count.to_frame(), popular_price.to_frame(), 
                               popular_purval.to_frame()],axis=1, sort=False)
popularity_analysis.columns=['Purchase Count', 'Item Price','Total Purchase Value']
popularity_sorted=popularity_analysis.sort_values(['Purchase Count', 'Item Price', 'Total Purchase Value'],
                                                  ascending=False).groupby('Purchase Count')
                                                  
popularity_sorted.head()
```
